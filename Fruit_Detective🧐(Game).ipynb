```python
import random
import time

# List of words to guess
word_list = ["apple", "mango", "banana", "cherry", "lichi", "pineapple", "watermelon",
             "avocado", "blueberry", "grape", "guava", "jackfruit", "kiwi", "orange",
             "papaya", "peach", "plum", "pomegranate", "strawberry", "raspberry"]
word_to_guess = random.choice(word_list)  # Randomly select a word
guessed_word = ["_"] * len(word_to_guess)  # Create a placeholder for the guessed word
attempts = 6  # Number of allowed wrong attempts
guessed_letters = []  # To keep track of guessed letters

print("Welcome to! \U0001F929 - Fruit Detective\U0001F9D0")
print()
time.sleep(1)
print("Rules:")
time.sleep(1)
print("-Guess the word, one letter at a time.")
print("-You have", attempts, "incorrect attempts allowed.")

print()
time.sleep(1)
print("Let's Begin 🎉🎊")

print()
print("The word is:", " ".join(guessed_word))
time.sleep(2)

while attempts > 0 and "_" in guessed_word:
    guess = input("\nEnter a letter: ").lower()

    # Check if input is valid
    if len(guess) != 1 or not guess.isalpha():
        print("Invalid input. Please enter a single letter.")
        continue

    # Check if letter is already guessed
    if guess in guessed_letters:
        print("You already guessed that letter!")
        continue

    guessed_letters.append(guess)

    # Check if the guess is in the word
    if guess in word_to_guess:
        print(f"Good job! {guess} is in the word.")
        # Reveal the guessed letter in the word
        for idx, letter in enumerate(word_to_guess):
            if letter == guess:
                guessed_word[idx] = guess
    else:
        attempts -= 1
        print(f"Wrong guess! {guess} is not in the word. You have {attempts} attempts left.")

    # Show the current state of the word
    print("Word: ", " ".join(guessed_word))
    print("Guessed letters:", ", ".join(guessed_letters))

# Check if the player won or lost
if "_" not in guessed_word:
    print("\nCongratulations! You guessed the word:\U0001F44F", word_to_guess)

else:
    print("\nGame over! The word was:\U0001F61E", word_to_guess)
    print("\nBetter Luck Next Time !\U0001F60A")
```
